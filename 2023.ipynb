```python
import pandas as pd

df2023 = pd.read_csv('FY2023_068_Contracts_Full_20240214_1.csv')
df2023
```

```python
relevant_columns = [
    'contract_transaction_unique_key', 'contract_award_unique_key', 'award_id_piid',
    'federal_action_obligation',
    'woman_owned_business', 'women_owned_small_business', 'economically_disadvantaged_women_owned_small_business',
    'joint_venture_women_owned_small_business', 'minority_owned_business', 'other_minority_owned_business',
    'emerging_small_business', 'minority_institution',
    'primary_place_of_performance_state_code', 'primary_place_of_performance_country_code',
    'action_date', 'action_date_fiscal_year',
    'recipient_name', 'recipient_duns', 'naics_code', 'naics_description', 'potential_total_value_of_award'
]

new_2023 = df2023[relevant_columns]

# Display the first few rows of the filtered DataFrame to ensure it contains only the relevant columns
new_2023.head()
```

```python
# AWARD DISTRIBUTION ANALYSIS
# Summarize the top NAICS codes represented in the dataset
naics_summary = new_2023['naics_description'].value_counts().head(5)  # Top 10 NAICS codes

# Display the summary
naics_summary

# The top NAICS codes represented in the dataset, which indicate the primary sectors receiving contracts
```

```python
# Geographic distribution of contracts by state code and country code
state_distribution = new_2023['primary_place_of_performance_state_code'].value_counts().head(5)

# Display the top 10 states and countries for contract performance location
print(state_distribution)

# INSIGHTS:
# The geographic distribution suggests that businesses located in or able to operate in the top states may have more opportunities for securing EPA contracts.
```

```python
ownership_types = [
    'women_owned_small_business', 'minority_owned_business'
]

# Initialize summary data structure
ownership_summary = {
    'Ownership Type': [],
    'Number of Contracts': [],
    'Total Potential Value of Award ($)': [],
    'Average Potential Value of Award ($)': []
}

# Filter data for each ownership type and calculate summary information
for ownership in ownership_types:
    ownership_data = new_2023[new_2023[ownership] == 't']
    total_value = ownership_data['potential_total_value_of_award'].sum()
    average_value = total_value / max(1, len(ownership_data))

    ownership_summary['Ownership Type'].append(ownership.replace('_', ' ').title())
    ownership_summary['Number of Contracts'].append(len(ownership_data))
    ownership_summary['Total Potential Value of Award ($)'].append(total_value)
    ownership_summary['Average Potential Value of Award ($)'].append(average_value)

# Convert to DataFrame for display
ownership_summary_df = pd.DataFrame(ownership_summary)

# Display the DataFrame
ownership_summary_df
```

```python
# Aggregate contract values by NAICS code
sector_value_summary = new_2023.groupby('naics_description').agg(
    Total_Potential_Value=('potential_total_value_of_award', 'sum'),
    Average_Potential_Value=('potential_total_value_of_award', 'mean'),
    Number_of_Contracts=('naics_code', 'count')
).reset_index()

# Sort by Total Potential Value to identify high-value sectors
sector_value_summary_sorted = sector_value_summary.sort_values(by='Total_Potential_Value', ascending=False)

# Display the top 10 NAICS codes by Total Potential Value
sector_value_summary_sorted.head(10)
```
