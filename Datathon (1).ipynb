```python
import pandas as pd
df2023 = pd.read_csv('FY2023_068_Contracts_Full_20240214_1.csv')
df2023

```

```python
df2023.info()
df2023.describe()
df2023.isna().sum()
# Dropped columns with completely missing values
df2023.drop(columns = ['recipient_duns', 'recipient_parent_duns', 'dod_claimant_program_code', 'dod_claimant_program_description', 'dod_acquisition_program_code',
                       'dod_acquisition_program_description', 'sea_transportation_code', 'sea_transportation'], axis = 1, inplace = True)
df2023
```

```python
relevant_columns = [
    'federal_action_obligation', 'potential_total_value_of_award',
    'base_and_exercised_options_value', 'current_total_value_of_award',
    'recipient_name', 'recipient_uei',
    'minority_owned_business', 'woman_owned_business', 'small_disadvantaged_business',
    'emerging_small_business', 'contracting_officers_determination_of_business_size',
    'contract_award_unique_key',
    'naics_description',
    'primary_place_of_performance_state_code', 'primary_place_of_performance_country_code',
    'period_of_performance_start_date', 'period_of_performance_current_end_date',
    'action_date', 'action_date_fiscal_year',
    'funding_agency_name', 'funding_sub_agency_name',
    'transaction_description'
]

df2023_1 = df2023[relevant_columns].copy()
df2023_1.info()
df2023_1
```

```python
df2023_1.isna().mean()
df2023_1.dropna(inplace = True)
```

```python
df2023_1.describe()
```

```python
# Preprocess the data: Convert 't'/'f' to True/False
df2023_1.replace({'t': True, 'f': False}, inplace=True)

# Group by NAICS description and aggregate
df2023_1.groupby('naics_description').agg({
    'woman_owned_business': 'sum',
    'small_disadvantaged_business': 'sum',
    'emerging_small_business': 'sum',
    'minority_owned_business': 'sum'
}).sort_values(by=[
    'woman_owned_business',
    'small_disadvantaged_business',
    'emerging_small_business',
    'minority_owned_business'
], ascending=False, inplace = True)
df2023_1.set_index('naics_description', inplace=True)

# Display the top 5 NAICS categories for these attributes
df2023_1
```

```python
# Calculate the distribution of each categorical variable
for column in ['minority_owned_business', 'woman_owned_business', 'small_disadvantaged_business', 'emerging_small_business']:
    distribution = df2023_1[column].value_counts()
    print(f"Distribution for {column}:\n{distribution}\n")
```

```python
# How the number of awards, potential total value, and average total value per award
# for minority-owned businesses have changed over time, specifically focusing on the fiscal year(s)

# Filter for minority-owned businesses
df2023_1 = df2023_1[df2023_1['minority_owned_business'] == True]

# Group by 'action_date_fiscal_year' to analyze trends
trend_analysis = df2023_1.groupby('action_date_fiscal_year').agg({
    'federal_action_obligation': ['count', 'sum', 'mean']
}).reset_index()

# Renaming columns for clarity
trend_analysis.columns = ['Fiscal Year', 'Number of Awards', 'Potential Total Value', 'Average Total Value per Award']

# Display the trend analysis
trend_analysis
```

```python
# This code performs a geographical analysis on contracts awarded to
# minority-owned businesses, grouped by country and state.
# It calculates the total and average federal obligation per award, formatting
# the average for readability. Results are sorted by the average total value
# per award to highlight regions with the most financially significant contracts on average.



# Group by both country and state codes for minority-owned businesses
geographical_trends = df2023_1[df2023_1['minority_owned_business']].groupby(
    ['primary_place_of_performance_country_code', 'primary_place_of_performance_state_code']
).agg({
    'federal_action_obligation': ['count', 'sum']
}).reset_index()

# Flatten the multi-level column names resulting from aggregation
geographical_trends.columns = ['Country Code', 'State Code', 'Number of Awards', 'Total Obligation']

# Calculate 'Average Obligation per Award'
geographical_trends['Average Obligation per Award'] = geographical_trends['Total Obligation'] / geographical_trends['Number of Awards']

# Sort by 'Average Obligation per Award' for insight into where contracts are most lucrative on average
geographical_trends.sort_values(by='Average Obligation per Award', ascending=False, inplace=True)

# Display the top entries for geographical trends by average obligation per award
geographical_trends.round(2)
```
